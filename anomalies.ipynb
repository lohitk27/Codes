```python
def alph_numeric(string):
    alphabet= ''.join(x for x in string if x.isalpha())
    number= ''.join(x for x in string if x.isdigit())
    
    sort_numbers= ''.join(sorted(number,reverse=True))
    sort_alpha = ''.join(sorted(alphabet,reverse=True))
    
    comb_string=''
    i,j=0,0
    for char in string:
        if char.isalpha():
            comb_string += sort_alpha[i]
            i +=1
        elif char.isdigit():
            comb_string += sort_numbers[j]
            j +=1
        else:
            comb_string +=char
    return comb_string

string="187A3b, 5c, 7de84!$"
sam=alph_numeric(string)
print(sam)
```

```python
from datetime import datetime, timedelta

def detect_anomalies(log_file_path):
    with open(log_file_path, 'r') as file:
        lines = file.readlines()
    suspicious_indices = []
    for i in range(len(lines)):
        line = lines[i]
        if is_anomalous(lines, i):
            suspicious_indices.append(i)
    suspicious_blocks = []
    for index in suspicious_indices:
        start_index = max(0, index - 2)
        end_index = min(len(lines), index + 4)
        block = lines[start_index:end_index]
        suspicious_blocks.append(block)
    return suspicious_blocks

def is_anomalous(lines, index):
    if index == 0 or index == len(lines) - 1:
        return False

    current_time = parse_timestamp(lines[index].split(" - ")[0])
    #print(current_time)
    previous_time = parse_timestamp(lines[index - 1].split(" - ")[0])
    #print(previous_time)
    next_time = parse_timestamp(lines[index + 1].split(" - ")[0])
    #print(next_time)
    #current log line has a significant deviation in timestamp compared to adjacent lines
    if (current_time - previous_time).total_seconds() > 10 or (next_time - current_time).total_seconds() > 30:
        return True
    else:
        return False

def parse_timestamp(timestamp_str):
    return datetime.strptime(timestamp_str, "%d %b %Y %H:%M:%S.%f")


log_file_path = "/Users/lohitkumar/Downloads/corelog_registration.txt"
suspicious_blocks = detect_anomalies(log_file_path)
for block in suspicious_blocks:
    print("".join(block))
    print("++++++++======\n")
```
